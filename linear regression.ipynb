# Trying to see if there is a linear relationship between test scores and number of hours studied

Loss Function: Mean Squared Error -> 
### 1/n * Σ((y_actual - (mx + b))^2)

Derivative of loss Function with respect to m -> 
### -2/n * Σ((y_actual - (mx + b)) * (x)) = D_m

Derivative of loss function with respect to b -> 
### -2/n * Σ((y_actual - (mx + b))) = D_b

Learning rate = L
### m step size -> m_new = m_current - L * D_m
### b step size -> b_new = b_current - L * D_b

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Sets size of graph
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
# Draw scatterplot of current data without best fit line
data = pd.read_csv('https://raw.githubusercontent.com/llSourcell/linear_regression_live/adc205e966e6efc5ec4794b2b645e29bb2c6b0c3/data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X,Y)
plt.show()
```

```python
#Building the model
m = 0
b = 0
L = 0.001
epochs = 1000
n = len(X)

for i in range(n):
    Y_pred = m*X + b
    D_m = (-2/n) * sum(X * (Y - Y_pred))
    D_b = (-2/n) * sum(Y - Y_pred)
    m = m - L*D_m
    b = b - L*D_b
```

```python
# Making Predictions
Y_pred = m*X + b

plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y), max(Y)], color = 'red')
plt.show()
```
